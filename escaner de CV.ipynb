```python
import docx2txt
```

```python
#Creamos una variable para que podamos verificar la informacion de nuestro CV o resume
CV = docx2txt.process("practicaCV.docx")
#ahora imprimos nuestro CV
print(CV)
```

```python
#ahora creamos una variable para la descripcion de trabajo
job_description=docx2txt.process("job_description.docx")
print(job_description)
```

```python
# una lista de texto
text = [CV,job_description]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#imprimir las puntuaciones similares
print("puntuaciones similares:")
print(cosine_similarity(count_matrix))
```

```python
#ahora conseguiremos el porcentaje de coincidencias
#match percentage
matchPercentage = cosine_similarity(count_matrix)[0][1] *100
mathcPercentage = round(matchPercentage, 2) # redondea a dos lugares decimales
print("tu partido de curriculum " + str(matchPercentage)+ "%de la descripcion de trabajo.")
```
